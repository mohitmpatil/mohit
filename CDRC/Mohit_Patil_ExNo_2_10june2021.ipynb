```python
#using PAndas read csv file
import pandas as pd
df = pd.read_csv(r"Gaussian Random number 8260 - Sheet1.csv")

# Create a DataFrame       

dataFrame = pd.DataFrame(df);
print("Contents of the DataFrame:");
print(dataFrame); 

# Convert the DataFrame to Series
dictionaryObject = dataFrame.to_dict();
print("DataFrame as a dictionary:");
print(dictionaryObject);
```

```python
# Import json module
import json


# Convert dictionary into json object without indentation
json_data = json.dumps(dictionaryObject)
# print json data
print("JSON Output without indentation: \n",json_data, "\n")

# Convert dictionary into json object with indentation
json_data = json.dumps(dictionaryObject,indent=3)
# print json data with indentation
print("Sorted JSON Output with indentation: \n", json_data)
```

```python
# Set the json filename
jsonFile = 'Set_1.json'
# Open a json file for writing json data
with open(jsonFile, 'w') as fileHandler1:
  json.dump(dictionaryObject, fileHandler1, indent = 2)

# Open a json file for reading
fileHandler2 = open(jsonFile)
print("The content of the JSON file: \n", fileHandler2.read())
```
