```python
import pandas as pd
df1 = pd.read_csv(r"Gaussian Random number 8260 - Sheet2.csv")
df2 = pd.read_csv(r"Gaussian Random number 8260 - Sheet1.csv")
res1 = pd.concat([df1, df2], axis=1, join='inner')

print("\n----------- Calculate Mean -----------\n")
mean=res1.mean()
print(mean)
 
print("\n----------- Calculate Median -----------\n")
med=res1.median()
print(med)
 
print("\n----------- Calculate Mode -----------\n")
mode=res1.mode()
print(mode)

print("\n----------- Calculate Stanadard Deviation -----------\n")
std=res1.std()
print(std)

# Convert the DataFrame to Series
dictmean = mean.to_dict();
dictmode = mode.to_dict();
dictmed = med.to_dict();
dictstd = std.to_dict();
print("Dictionary:");
Dictionar=(dictmean,dictmed,dictstd,dictmode)
print(Dictionar)
```

```python
import json
# Set the json filename
jsonFile = 'final.json'
# Open a json file for writing json data
with open(jsonFile, 'w') as fileHandler1:
  json.dump(Dictionar, fileHandler1, indent = 2)

# Open a json file for reading
fileHandler2 = open(jsonFile)
print("The content of the JSON file: \n", fileHandler2.read())
```
